```python
content = open("input.txt", "r", encoding = "utf-8")
print(content.read())
content.close()
```

```python
with open("input.txt", "r", encoding="utf-8") as f_in:   #mở file nguồn để đọc
    content = f_in.read()                                #đọc toàn bộ nội dung
with open("copy.txt", "w", encoding="utf-8") as f_out:   #mở or tạo file đích để ghi
    f_out.write(content)                                 #đoạn này đểghi nội dung sang file mới
```

```python
with open("copy.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()  #này là để đọc cái file input đó
selected = lines[:5]# lấy 5 dòng đầu

#ghi sang file output.txt với cái họ tên
with open("output.txt", "w", encoding="utf-8") as f:
    f.writelines(selected) # ghi 5 dòng đầu của file cũ vô file mới
    f.write("Phạm Duy Anh\n")
```
